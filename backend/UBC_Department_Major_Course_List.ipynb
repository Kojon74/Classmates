```
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame

URL = 'https://courses.students.ubc.ca/cs/courseschedule?pname=subjarea&tname=subj-all-departments'
page = requests.get(URL)

soup = BeautifulSoup(page.content, 'html.parser')
```

```
results = soup.find(id='mainTable')

majors1=[]
for link in results.find_all("td"):
    print("Inner Text: {}".format(link.text))
    majors1.append(link.text)
    print("href: {}".format(link.get("href")))

majors = [] 
for ele in majors1: 
    if ele.strip(): 
        majors.append(ele) 
print(majors)
```

```
len(majors)
```

```
# Department & Major Code dataframe
ubc_majors=[]
ubc_department=[]
k=0
for i in range(0, int(len(majors)/3)-1):
  ubc_majors.append(majors[3*i])
  ubc_department.append(majors[3*i+2])

ubc_list=[]
for i in range(len(ubc_department)):
  ubc_list.append([ubc_department[i]] + [ubc_majors[i]])
```

```
# Department & Major Code & Course IDs Dataframe

k=0
for link in results.find_all("a"):
  # redirect to course website
  URL='https://courses.students.ubc.ca'+str(link.get("href"))
  page = requests.get(URL)
  soup = BeautifulSoup(page.content, 'lxml')
  # get the main table of course id list
  coursetable = soup.find(id="mainTable")
  # get different ids in list
  courseid=[]
  for ids in coursetable.find_all("a"):
    courseid.append(ids.text)

  courseids = [] 
  for ele in courseid: 
      if ele.strip(): 
        courseids.append(ele) 
  print(courseids)
  
  ubc_list[k].append(courseids)
  k+=1
```

```
print(ubc_list)
```

```
df = DataFrame (ubc_list,columns=['Department','Major_Code', 'Course_IDs'])
df.head()
```
