```
import requests
from bs4 import BeautifulSoup
from pandas import DataFrame
```

```
# UofC
URL_uc = 'https://www.ucalgary.ca/pubs/calendar/current/course-by-faculty.html'
page_uc = requests.get(URL_uc)
soup_uc = BeautifulSoup(page_uc.content, 'html.parser')
results_uc = soup_uc.find(id='ctl00_ctl00_pageContent')

## Find the major name
majors1_uc=[]
for link in results_uc.find_all("a"):
    print("Inner Text: {}".format(link.text))
    majors1_uc.append(link.text)
    print("href: {}".format(link.get("href")))

majors_uc = [] 
for ele in majors1_uc: 
    if ele.strip(): 
        majors_uc.append(ele) 
print(majors_uc)

## Convert to a dataframe
df_uc = DataFrame (majors_uc,columns=['Majors'])
df_uc.head()
df_uc.to_csv(r'majors_data_uofc.csv')
```

```
# UBC
URL_ubc = 'https://courses.students.ubc.ca/cs/courseschedule?pname=subjarea&tname=subj-all-departments'
page_ubc = requests.get(URL_ubc)

soup_ubc = BeautifulSoup(page_ubc.content, 'html.parser')
results_ubc = soup_ubc.find(id='mainTable')

## Find the major name
majors1_ubc=[]
for link in results_ubc.find_all("td"):
    print("Inner Text: {}".format(link.text))
    majors1_ubc.append(link.text)
    print("href: {}".format(link.get("href")))

majors_ubc = [] 
for ele in majors1_ubc: 
    if ele.strip(): 
        majors_ubc.append(ele) 
print(majors_ubc)

## Convert to a dataframe
majors_ubc_abbrv=[]
k=0
for i in range(0, int(len(majors_ubc)/3)-1):
  majors_ubc_abbrv.append(majors_ubc[3*i])

df_ubc = DataFrame (majors_ubc_abbrv,columns=['Majors'])
df_ubc.head()
df_ubc.to_csv(r'majors_data_ubc.csv')
```
